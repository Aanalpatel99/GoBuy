```python
import cv2

def openCam():
    # Open the camera
    cap = cv2.VideoCapture(0)
    cap.set(3, 640)  # Set width
    cap.set(4, 480)  # Set height

    print("Press 'q' to quit the barcode scanning")

    while True:
        success, frame = cap.read()
        if not success:
            break

        # Draw a square on the frame as a guide box for barcode placement
        # Coordinates for the rectangle (centered in the frame)
        start_point = (220, 140)  # Top-left corner (x, y)
        end_point = (420, 340)    # Bottom-right corner (x, y)
        color = (0, 255, 0)       # Green color in BGR
        thickness = 2             # Thickness of the rectangle border

        # Draw the rectangle on the frame
        cv2.rectangle(frame, start_point, end_point, color, thickness)

        # Decode QR code using OpenCV
        flag = decode_qr_code(frame)
        if flag:
            break
        # Display the frame
        cv2.imshow('Barcode Scanner', frame)

        # Break loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the camera and close windows
    cap.release()
    cv2.destroyAllWindows()

def decode_qr_code(frame):
    detector = cv2.barcode.BarcodeDetector()
    data, bbox, _ = detector.detectAndDecode(frame)
    if data:
        print('Bar code detected', data)  #("Decoded data:", data)
        return 1
    else:
        print("No bar code found")  #('Bar code detected: Item name: chips')
# Run the function
openCam()
```
